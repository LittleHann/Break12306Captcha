```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse_and_plot_log(log_name):
    with open(log_name) as log_reader:
        losses = []
        iterations = []
        for line in log_reader:
            i_loss = line.find('loss = ')
            if i_loss > 0:
                iterations.append(len(losses) * 5)
                losses.append(float(line[i_loss + 7:]))
    plt.plot(iterations, losses)
```

```python
fname = '../../log/margin_0.5_hard_0.7/training.log'
parse_and_plot_log(fname)
```

```python
fname = '../../log/margin_0.5_hard_0.5/training.log'
parse_and_plot_log(fname)
```
