```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

%matplotlib inline


ROW = 4
COL = 4
color_im = Image.open('../data/captcha_0.jpg').crop((10, 40, 73, 105))
gray_im = color_im.convert('L')
color_arr = np.array(color_im)
gray_arr = np.array(gray_im)
resized_color_arr = np.array(color_im.resize((8,8)))
resized_gray_arr = np.array(color_im.resize((6, 6)).convert("L"))
binary_gray_arr = resized_gray_arr > np.mean(resized_gray_arr)
binary_red_arr = resized_color_arr[:, :, 0] > np.mean(resized_color_arr[:, :, 0])
binary_green_arr = resized_color_arr[:, :, 1] > np.mean(resized_color_arr[:, :, 1])
binary_blue_arr = resized_color_arr[:, :, 2] > np.mean(resized_color_arr[:, :, 2])
binary_gray_vec = binary_gray_arr.reshape((1, 36))
binary_gray_vec = binary_gray_vec.astype(int)
binary_red_vec = binary_red_arr.reshape((1, 64))
binary_green_vec = binary_green_arr.reshape((1, 64))
binary_blue_vec = binary_blue_arr.reshape((1, 64))   

plt.figure(frameon='true')
plt.subplot(ROW, COL, 1)
plt.imshow(color_arr)
plt.axis('off')
plt.subplot(ROW, COL, 3)
plt.axis('off')
plt.imshow(gray_arr, cmap=plt.get_cmap('gray'), interpolation="nearest")
plt.subplot(ROW, COL, 2)
plt.axis('off')
plt.imshow(resized_color_arr)
plt.axis('off')
plt.subplot(ROW, COL, 4)
plt.imshow(resized_gray_arr, cmap=plt.get_cmap('gray'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 6)
plt.imshow(resized_color_arr[:, :, 0], cmap=plt.get_cmap('Reds_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 7)
plt.imshow(resized_color_arr[:, :, 1], cmap=plt.get_cmap('Greens_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 8)
plt.imshow(resized_color_arr[:, :, 2], cmap=plt.get_cmap('Blues_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 5)
plt.imshow(resized_gray_arr, cmap=plt.get_cmap('gray'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 10)
plt.imshow(binary_red_arr, cmap=plt.get_cmap('Reds_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 11)
plt.imshow(binary_green_arr, cmap=plt.get_cmap('Greens_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 12)
plt.imshow(binary_blue_arr, cmap=plt.get_cmap('Blues_r'), interpolation="nearest")
plt.axis('off')
plt.subplot(ROW, COL, 9)
plt.imshow(binary_gray_arr, cmap=plt.get_cmap('gray'), interpolation="nearest")
plt.axis('off')
# plt.subplot(ROW, COL, 13)
# plt.gca().set_aspect(1./64, adjustable='box')
# plt.imshow(binary_red_vec, cmap=plt.get_cmap('Reds_r'), interpolation="nearest")


```
