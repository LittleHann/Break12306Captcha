```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse_and_plot_log(log_name):
    with open(log_name) as log_reader:
        losses = []
        iterations = []
        for line in log_reader:
            i_loss = line.find('loss = ')
            if i_loss > 0:
                iterations.append(len(losses) * 5)
                losses.append(float(line[i_loss + 7:]))
    plt.plot(iterations, losses)
```

```python
margin01_fname = '../../log/margin_01.log'
parse_and_plot_log(margin01_fname)
```

```python
margin05_fname = '../../log/margin_05.log'
parse_and_plot_log(margin05_fname)
```
