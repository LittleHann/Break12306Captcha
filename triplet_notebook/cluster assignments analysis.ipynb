```python
from collections import Counter

clusters = []

with open('../data/200_cluster.txt') as reader:
    for line in reader:
        cur_assignments = line.strip().split()
        num_images = len(cur_assignments)
        
        cur_counter = Counter(cur_assignments)
        for k, v in cur_counter.iteritems():
            cur_counter[k] = v / float(num_images)
            
        cur_cluster = sorted(list(cur_counter.iteritems()), key=lambda t: t[1], reverse=True)
        
        clusters.append(cur_cluster[:3])

clusters.sort(key=lambda c: c[0][1], reverse=True)

for cur_cluster in clusters:
    for label, ratio in cur_cluster:
        print label, ratio,
    else:
        print
```

```python
from collections import Counter 

all_labels = []
with open('../data/400_cluster.txt') as reader:
    for line in reader:
        all_labels.extend(line.strip().split())
        
num_images = len(all_labels)
label_counter = Counter(all_labels)
```

```python
ratios = []
for label, count in label_counter.most_common():
    ratio = count / float(num_images)
    ratios.append(ratio)
#     print label, count, ratio
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.hist(np.array(ratios))
plt.show()
```
